```python
graphql_endpoint = 'https://www.instagram.com/graphql/query/'
graphql_followers = (
    graphql_endpoint + '?query_hash=37479f2b8209594dde7facb0d904896a')

all_followers = []

variables = {}
variables['id'] = '22597626903' # 굽네치킨
# variables['id'] = '3024418777' # kent_busan
# variables['id'] = '7429511270' # alsoj_pic (600명)
variables['first'] = 50

# get follower and user loop
sc_rolled=0
try:
    has_next_data = True

    url = (
        '{}&variables={}'
        .format(graphql_followers, str(json.dumps(variables)))
    )
    sleep(2)
    browser.get(url)

    # fetch all user while still has data
    while has_next_data:
        pre = browser.find_element_by_tag_name("pre").text
        data = json.loads(pre)['data']

        # get followers
        page_info = (
            data['user']['edge_followed_by']['page_info'])
        edges = data['user']['edge_followed_by']['edges']
        for user in edges:
            all_followers.append(user['node']['username'])
#             sheet.append(list(user['node']['username']))
            print("팔로워 추출 중 : ", user['node']['username'])

        has_next_data = page_info['has_next_page']
        if has_next_data:
            variables['after'] = page_info['end_cursor']

            url = (
                '{}&variables={}'
                .format(
                    graphql_followers, str(json.dumps(variables)))
            )
            sleep(2)
            browser.get(url)
            sc_rolled+=1
            
#             try:
#                 print("sc_rolled ::: " + str(sc_rolled))
#                 print("all_followers cnt : ", len(all_followers))
#                 filename = '{}watchers_queries.json'.format(logfolder)
#                 sc_rolled_date = datetime.date.today().strftime('%Y-%m-%d')
#                 if not os.path.isfile(filename):
#                     with open(filename,'w+') as sc_rolled_data:
#                         json.dump({}, sc_rolled_data)
#                         sc_rolled_data.close()
#                 """Loads the existing data"""
#                 with open(filename) as sc_rolled_data:
#                     sc_rolled_all = json.load(sc_rolled_data)
#                 sc_rolled_all[sc_rolled_date] = sc_rolled_all.get(sc_rolled_date, 0) + 1
#                 """Dumps the current data"""
#                 with open(filename, 'w') as sc_rolled_data:                     
#                     json.dump(sc_rolled_all, sc_rolled_data)
#                     sc_rolled_data.close()
#             except:
#                 raise

            if sc_rolled > 91:
                print("일시 휴식!!")
                print("all_followers cnt : ", len(all_followers))
                sleep(600)
                sc_rolled = 0
        
except BaseException:
#     logger.info("Unable to get followers information\n")
    raise

my_watchers = list(set(all_followers))
sleep(566)
```
